```python
import shutil
import subprocess

# ì‹¤ì‹œê°„ ì¶œë ¥ì„ ê°•ì œë¡œ í™”ë©´ì— ë¿Œë ¤ì£¼ëŠ” í—¬í¼ í•¨ìˆ˜
def run_live_output(command):
    # í”„ë¡œì„¸ìŠ¤ ì‹¤í–‰ (stdoutì„ íŒŒì´í”„ë¡œ ì—°ê²°)
    process = subprocess.Popen(
        command,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,  # ì—ëŸ¬ ë©”ì‹œì§€ë„ ê°™ì´ ì¶œë ¥
        text=True,                 # í…ìŠ¤íŠ¸ ëª¨ë“œë¡œ ì½ê¸°
        bufsize=1                  # ë¼ì¸ ë²„í¼ë§
    )

    # í”„ë¡œì„¸ìŠ¤ê°€ ë±‰ëŠ” ë¡œê·¸ë¥¼ í•œ ì¤„ì”© ì‹¤ì‹œê°„ìœ¼ë¡œ print
    for line in process.stdout:
        print(line, end='')

    # í”„ë¡œì„¸ìŠ¤ ì¢…ë£Œ ëŒ€ê¸° ë° ì—ëŸ¬ ì²´í¬
    process.wait()
    if process.returncode != 0:
        raise subprocess.CalledProcessError(process.returncode, command)

# --- ë©”ì¸ ì‹¤í–‰ ë¡œì§ ---

# 1. ì„¤ì • íŒŒì¼ ë³µì‚¬
print("=== [1] MNIST_MLP ì„¤ì •ì„ ì ìš©í•©ë‹ˆë‹¤... ===")
shutil.copy('config_mnist_mlp.yaml', 'config.yaml')

# 2. train.py ì‹¤í–‰
print("\n=== [2] ëª¨ë¸ í•™ìŠµ ì‹œì‘! ===")
run_live_output([sys.executable, "-u", "train.py"])

# 3. eval.py ì‹¤í–‰
print("\n=== [3] ëª¨ë¸ í‰ê°€ ì‹œì‘! ===")
run_live_output([sys.executable, "-u", "eval.py"])
```
