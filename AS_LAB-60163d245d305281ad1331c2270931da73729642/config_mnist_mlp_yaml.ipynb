```python
# 파일 이름을 명시적으로 지정
FILE_NAME = "config_mnist_mlp.yaml"
# 전체 경로 사용
SAVE_PATH = f"/content/drive/MyDrive/AS_LAB/{FILE_NAME}"
# 작성 시 변수 대신 직접 확인하며 실행
```

```python
config_text = """
project_name: "MNIST_MLP_Baseline"

data:
  dataset_name: "mnist"
  batch_size: 64
  num_workers: 0
  augmentation:
    use_aug: False
    rotation: 10
    flip_prob: 0.5

model:
  type: "mlp"
  input_size: 784
  hidden_size: 256
  num_classes: 10
  dropout_rate: 0.0

train:
  num_epochs: 30
  learning_rate: 0.001
  seed: 42
  save_path: "best_mnist_mlp.pth"

  #고급 학습 설정
  patience: 10         # Early Stopping 기다리는 횟수
  optimizer: "adam"    # adam 또는 sgd
  weight_decay: 0.0    # L2 규제 (0.0001 등)

  #스케줄러 설정
  use_scheduler: False
  scheduler_step: 10
  scheduler_gamma: 0.1
"""

# 파일 저장
with open(SAVE_PATH, 'w', encoding='utf-8') as f:
    f.write(config_text.strip())

print(f"설정 파일이 성공적으로 업데이트되었습니다: {SAVE_PATH}")
```
