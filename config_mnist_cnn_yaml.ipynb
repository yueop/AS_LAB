```python
from google.colab import drive
import os

drive.mount('/content/drive')
Project_Path = ('/content/drive/MyDrive/AS_LAB')
if not os.path.exists(Project_Path):
    os.makedirs(Project_Path)

os.chdir(Project_Path)
```

```python
FILE_NAME = 'config_mnist_cnn.yaml'
SAVE_PATH = f"/content/drive/MyDrive/AS_LAB/{FILE_NAME}"
```

```python
config_text = """
project_name: "MNIST_CNN"

data:
    dataset_name: "mnist"
    batch_size: 64
    num_workers: 0
    augmentation:
        use_aug: False
        rotation: 10
        flip_prob: 0.0

model:
    type: "cnn"
    input_size: 784
    hidden_size: 256
    num_classes: 10
    dropout_rate: 0.0

train:
    num_epochs: 30
    learning_rate: 0.01
    seed: 42
    save_path: "best_mnist_cnn.pth"

    patience: 3
    optimizer: "adam"
    weight_decay: 0.0

    use_scheduler: False
    scheduler_step: 10
    scheduler_gamma: 0.1
"""

with open(SAVE_PATH, 'w', encoding='utf-8') as f:
    f.write(config_text.strip())

print(f"설정 파일이 성공적으로 업데이트되었습니다: {SAVE_PATH}")
```
